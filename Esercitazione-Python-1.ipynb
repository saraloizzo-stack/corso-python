**Esercizio:** scrivi un programma che chieda all'utente un numero intero positivo N e stampi tutti i numeri primi compresi tra 2 e N.

**Regole e vincoli:**


*   Un numero è primo se è divisibile solo per 1 e per se stesso.
*   Non puoi usare liste o altre strutture dati.

*   Devi usare un ciclo dentro l'altro (nested loops).

```python
N=int(input("inserisci un numero: ")) #chiede all'utente un numero intero

for i in range (2,N+1): # ciclo sui numeri da 2 a N
  primo=True #i primo
  for j in range (2,1):
    if i %j==0: #se i è divisibile per j
      primo=False # i non è primo
  if primo: #se non è divisibile allora
    print(i) #stampa il numero
```
